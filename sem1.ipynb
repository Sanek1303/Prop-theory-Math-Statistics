**Математическая статистика** — раздел математики, разрабатывающий методы регистрации, описания и анализа данных наблюдений и экспериментов с целью построения вероятностных моделей массовых случайных явлений.

**Определение № 1** Совокупность наблюдаемых случайных величин $X = (X_1,..., X_n)$   называется _(случайной) выборкой_, сами величины $X_k$ - _элементами выборки_, а их число $n$ - ее _объемом_.

**Определение № 2** Множество $X = \{x = (x_1,...,x_n)\}$ всех возможных реализаций выборки $X = (X_1,...,X_n)$ называется _выборочным пространством_.

**Определение № 3**. Пусть $F_x(x_1,...,x_n) = P\{X_1< x_1,...,X_n < x_n\}$ – функция распределения выборки (возможно неизвестная), принадлежащая известному классу распределений $F = \{F(x_1,...,x_n)\}$. Пара $(X, F)$ называется статистической моделью. Иногда первый элемент пары опускается _(статистическая модель $F$)_.

Математическая статистика решает (в рамках заданной модели $F$) задачи уточнения / определения различных свойств истинного распределения $F_x$ по результатам проводимых наблюдений (по выборке $X$).

**Определение № 4**. Если $X = (X_1,...,X_n)$ — НОРСВ, имеющие распределение $L(Y)$ (распределенная по закону $L(Y)$), то говорят, что $X$  является выборкой из распределения $L(Y)$. В этом случае (модели повторных независимых наблюдений) в статистической модели $F = \{F_Y\}$ указывается распределение только одного элемента выборки.

**Определение № 5** Если $F = \{F(x, \theta), \theta \in \Theta\}$, то такая модель называется параметрической,
а множество $\Theta$ возможных значений параметра $\theta$  — _параметрическим множеством_.

Во многих задачах математической статистики рассматриваются последовательности случайных величин $\eta_n$, сходящиеся в том или ином смысле к некоторому пределу $\eta$  — случайной величине или константе, когда $\eta \rightarrow \infty$. Обычно рассматривают сходимость по вероятности и по распределению (иногда – почти наверное).

**Определение № 6** Если $X = (X_1,...,X_n)$ — выборка из некоторого распределения $L(Y)$, то $F_Y=F(x)$ называют теоретической функцией распределения, а 
$$F_n(x) = \frac{\mu_n(x)}{n} = \frac{1}{n} \sum_{k = 1}^{n} I(x - X_k)$$


— эмпирической (выборочной) функцией распределения. Здесь $\mu_n(x)$ – число элементов выборки, попавших строго левее $x$ $(X_k < x)$, а $I(x - X_k)$ – единичная ступенчатая функция, непрерывная слева:

\begin{equation*}
I(x) = 
 \begin{cases}
   0, if  x \leq 0,\\
   1, if x > 0.
 \end{cases}
\end{equation*}



**Теорема 1 (Бернулли)**. При многократном повторении случайного эксперимента с двумя исходами относительная частота успехов сходится к вероятности успеха в одном испытании по вероятности.

**Теорема 2 (ЗБЧ в форме Чебышева)**. Пусть $\xi_1, \xi_2, . . .$ – последовательность независимых одинаково распределенных случайных величин, причем $E\xi_1 = a$ и $D\xi_1 = \sigma^2 < \infty$. Тогда для этой последовательности выполняется закон больших чисел, то есть: 
$$ \frac{1}{n} \sum_{k=1}^{n} \xi_k \xrightarrow{p} a$$ при $$ n \rightarrow \infty$$

**Теорема 3 (ЗБЧ в форме Хинчина)**. Пусть $\xi_1, \xi_2, . . .$ – последовательность независимых одинаково распределенных случайных величин с конечным математическим ожиданием. Тогда к последовательности применим закон больших чисел.

**Теорема 4 (УЗБЧ в форме Колмогорова)**. Пусть $\xi_1, \xi_2, . . .$ – последовательность независимых одинаково распределенных случайных величин. Для выполнения усиленного закона больших чисел для этой последовательности, т.е.
$$ \frac{1}{n} \big(\sum_{k = 1}^{n}\xi_k -  \sum_{k = 1}^{n} E\xi_k \big) \xrightarrow{} 0$$
 

необходимо и достаточно существование у случайной величины $\xi_1$ конечного математического ожидания.

**Определение 8**. Пусть $X = (X_1,...,X_n)$  — выборка из некоторого распределения. Определим разбиение числовой прямой $-\infty < a_0 < a_1 < ... < a_k <  +\infty$. Пусть  – $\Delta_1 = [a_0, a_1], \Delta_i = [a_{i-1}, a_i], i = 2,...k$ - разряды,
$$ n_i = \sum_{j = 1}^{n} (I_{\Delta_i}(X_j)), i = 1,2,...k $$
$n_i$ - число элементов выборки, попавших в i-й разряд. Тогда кусочно - постоянная функция $h: R \rightarrow R$, имеющая вид:
$$ h(x) = \sum_{i = 1}^{k} \frac{n_i}{n(a_i - a_{i-1})} I_{\Delta_i}(x)$$
$h(x)$ - называется гистограммой выборки

### №1

С помощью R[0,1] реализовать датчик распределения Bi(5, 1/2). По смоделированным выборкам длины 100 и 100 000 построить на одном рисунке графики ряда частот и теоретический график. По выборкам вычислить выборочные среднее и дисперсию, сравнить их с теоретическими значениями.

$F(x) = P(X < x) = \sum_{k \leq x} C_n^k p^k \cdot (1-p)^{n-k} $, где n это количество испытаний Бернулли

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

Идея: все вероятности кладем на [0, 1]. Генерируем с.в. из R[0, 1], смотрим какой вероятности соответствует данная с.в. По номеру вероятности восстанавливаем величину k.



```python
import itertools
n = 5
p = 0.5

def factorial(n):
    if (n <= 1):
        return 1
    else:
        return n*factorial(n-1)
    
def C(n,k):
    return factorial(n) / (factorial(n-k)*factorial(k))

def Bi(n, p):
    return np.array([C(n, i)*p**(i)*(1-p)**(n-i) for i in range(n+1)])

x = np.arange(n + 1)
y = Bi(n, p)
print(np.sum(y))
plt.plot(x, y)
```

```python
n1 = 100
n2 = 100000

def CumBi(n, k):
    y = Bi(n, k)
    cum_y = [None]*len(y)
    cum_y[0] = y[0]
    for i in range(1, len(y)):
        cum_y[i] = y[i]
        cum_y[i] += cum_y[i-1]
    return cum_y

def BiGenerator(n, k):
    cum_y = CumBi(n, k)
    x = np.random.uniform()
    for k in range(len(cum_y)):
        if (cum_y[k] > x):
            return k
        
def Frequencies(n, k, length):
    y = np.sort(np.array([BiGenerator(n, k) for i in range(length)])) 
    res = y.copy()
    y = np.bincount(y) / length    
    return y, res

print("E_theor = ", 5*0.5)
print("D_theor = ", 5*0.5*0.5)

x_theor = np.arange(5 + 1)
y_theor = Bi(5, 0.5)
plt.plot(x_theor, y_theor)

x = np.arange(6)
y, res = Frequencies(5, 0.5, n1)
plt.plot(x, y)
plt.scatter(x,y, c= 'r')
plt.show()



print("E_100 = ", np.sum(res)/len(res))
print("D_100 = ", np.sum(res*res)/len(res) - (np.sum(res)/len(res))**2)

x = np.arange(6)
y, res = Frequencies(5, 0.5, n2)
plt.plot(x_theor, y_theor)
plt.plot(x, y)
plt.scatter(x,y, c= 'r')
plt.show()

print("E_100_000 = ", np.sum(res)/len(res))
print("D_100_000 = ", np.sum(res*res)/len(res) - (np.sum(res)/len(res))**2)
```

На одном рисунке построить гистограммы смоделированных выборок (число и положение разрядов выбрать самостоятельно, но осмысленно!) и теоретическую плотность распределения.
```python
n1 = 100
n2 = 100000

def f(x):
    return 1/(np.pi * (1 + x*x))
def F(x):
    return 1/np.pi*np.arctan(x) + 1/2
    
def inv_F(x):
    # x from range[0,1]
    return np.tan((x-1/2)*np.pi)

def Emp_dist_func(X, x):
    #длина выборки
    n = len(X)
    return np.sum(X < x) / n
    
    

x = np.linspace(-10, 10, 100)
y = F(x)


unif_1 = np.array([np.random.uniform() for i in range(n1)])
kosh_1 = inv_F(unif_1)
y_1 = np.array([Emp_dist_func(kosh_1, item) for item in x])
plt.figure(figsize = (16, 9))
plt.plot(x, y)
plt.plot(x, y_1)
plt.show()



unif_2 = np.array([np.random.uniform() for i in range(n2)])
kosh_2 = inv_F(unif_2)
y_2 = np.array([Emp_dist_func(kosh_2, item) for item in x])
plt.figure(figsize = (16, 9))
plt.plot(x, y)
plt.plot(x, y_2)
plt.show()

    
    

    
```

Строим плотность распределения и гистограммы

```python
def hist(x, X, step = 2):
    delts = []
    intervals = []
    h_x = []
    n_i = []
    n = len(X) # размер выборки
    for i in range(step, len(x), step):
        delts.append(x[i] - x[i-step])
        intervals.append((x[i - step], x[i]))
    for i,interval in enumerate(intervals):        
        n_i.append(np.sum((interval[0] < X)*(X <= interval[1])))
    n_i = np.array(n_i)
    delts = np.array(delts)
    def func(x, n, n_i, delts, intervals):
        res = 0
        for i in range(len(n_i)):
            res += n_i[i]/(n*delts[i]) * (intervals[i][0] < x)*(x <= intervals[i][1])
        return res
               
    for item in x:
        h_x.append(func(item, n, n_i, delts, intervals))           
    return np.array(h_x)  


x = np.linspace(-10, 10, 1000)

y = f(x)
plt.figure(figsize = (16, 9))
plt.plot(x, hist(x, kosh_1, 25))
plt.plot(x, y)
plt.show()


y = f(x)
plt.figure(figsize = (16, 9))
plt.plot(x, hist(x, kosh_2, 25))
plt.plot(x, y)
```

Разряды гистограмм выбирать двумя способами: 1) совпадающими с теоретическими разрядами распределения, 2) разбивающими теоретические разряды пополам. По выборкам вычислить выборочные среднее и дисперсию, сравнить их с теоретическими значениями.

```python
def f(x):
    if (x <= 0):
        return 0
    elif (0 < x <= 1):
        return 0.05
    elif (1 < x <= 2):
        return 0.15
    elif (2 < x <= 3):
        return 0.4
    elif (3 < x <= 4):
        return 0.3
    elif (4 < x <= 5):        
        return 0.1
    else : return 0
    
x = np.linspace(-10, 10, 1000)
y = np.array([f(item) for item in x])
plt.figure(figsize = (16, 9))
plt.plot(x, y)
plt.show()

P = np.array([0.05, 0.15, 0.4, 0.3, 0.1])
a = [0,1,2,3,4,5]

#прямая через точку (x0, y0) с наклоном k
def l(x, x0, y0, k):
    return (x - x0)*k + y0
    
def F(x, a, k):
    indx = len(a)
    for i in range(len(a)):
        if (x < a[i]):
            indx = i
            break
    if (indx <= 0):
        return 0
    if (indx > len(a)-1):
        return 1
    
    l_a = 0
    for i in range(1, indx+1):        
        if (i >= 1 and i < indx):            
            l_a = l(a[i],a[i-1], l_a, k[i - 1])
        elif (i == indx):
            return l(x, a[i-1], l_a, k[i - 1])
        
        
def F_inv(x, a, k):
    nodes = np.array([F(a[i], a, k) for i in range(len(a))])
    new_k = -1/k
    return -F(x, nodes, new_k)
    
        

x = np.linspace(-10, 10, 1000)
y = np.array([F(item,a , P) for item in x])
plt.figure(figsize = (16, 9))
plt.plot(x, y)
plt.show()

            
x = np.linspace(0.001, 0.999, 100)
y = np.array([F_inv(item,a , P) for item in x])
plt.figure(figsize = (16, 9))
plt.plot(x, y)
plt.show()
            
        
```

Вычислить выборочные среднее, медиану, дисперсию, коэффициент асимметрии и куртозис. Сравнить полученные оценки с теоретическими. 

```python
from scipy.special import erf
from scipy.stats import skew, kurtosis
```

```python
# по задаче № 9

n1 = 100
n2 = 100000

def GenN():
    y = np.array([np.random.uniform() for _ in range(12)])    
    return np.sum(y) - 6

norm_1 = np.array([ GenN() for _ in range(n1)])
norm_2 = np.array([GenN() for _ in range(n2)])

print("E_100 = ", norm_1.mean())
print("E_100_000 = ", norm_2.mean())
print("D_100 = ", (norm_1.std())**2)
print("D_100_000 = ", (norm_2.std())**2)
print("Med_100 = ", np.median(norm_1))
print("Med_100_000 = ", np.median(norm_2))
print("Skew_100 = ", skew(norm_1))
print("Skew_100_000 = ", skew(norm_2))
print("Kurtosis_100 = ", kurtosis(norm_1))
print("Kurtosis_100_00 = ", kurtosis(norm_2))




def f_std(x):
    return 1/np.sqrt(2*np.pi) * np.exp(-(x*x)/2)

def F_std(x):
    return 1/2*(1 + erf(x/np.sqrt(2)))
plt.figure(figsize = (16, 9))
x = np.linspace(-5, 5, 1000)
plt.plot(x, F_std(x), c = 'b')
y1 = np.array([Emp_dist_func(norm_1, item) for item in x])
plt.plot(x, y1, c = 'r')
y2 = np.array([Emp_dist_func(norm_2, item) for item in x])
plt.plot(x, y2, c =  'g')
plt.show()


plt.figure(figsize = (16, 9))
x = np.linspace(-5, 5, 1000)
plt.plot(x, f_std(x), c = 'b')
plt.plot(x, hist(x, norm_1, step = 25), c= 'r')
plt.plot(x, hist(x, norm_2, step = 25), c= 'g')
plt.show()
```
