# Метод наименьших квадратов

Рассмотрим следюущую схему наблюдений : 

(1) $$ y_{t_i}  = \sum_{k = 1}^{K} f_k(t_i)\beta_k + \varepsilon_i, i = 1,..., T \;$$   
где $f_k(t) \; (k = 1,..., K)$ - известные, детерминированные функции, $\beta_k \; (k = 1,...,K)$ - неизвестные детерминированные параметры, подлежащие оцениванию, $\varepsilon_i \; (i = 1,..., T)$ - независимые одинаково распределенные гауссовские случайные величины $N(0, \sigma^2)$.

Наблюдения (1) могут быть записаны в матричной форме:  
(1') $$ Y = FB + E $$  
где $Y = (y_{t_1}, ...y_{t_T})$, $E = (\varepsilon_1,...,\varepsilon_T)$, вектор регрессоров - $B = (\beta_1,...,\beta_K)$  
  
\begin{equation*}
F = \begin{bmatrix}
f_1(t_1) ... f_k(t_1) \\
  ............... \\
f_1(t_T) ... f_k(t_T)
\end{bmatrix} - T \times K 
\end{equation*}


$F$ - матрица наблюдений. Зависимость от вектора оцениваемых неизвестных параметров в (1') линейная, поэтому она носит назавние линейной регрессии. 




### Задача № 1
По наблюдениям $Y$ найти СК оптимальну линейную оценку вектора параметров $B$.   
Решение :  
Будем искать оптимальную оценку $\hat{B}$ в виде $\hat{B} = \alpha Y$, где $\alpha$ - матрица соответствующей размерности, подлежащая определению.  
  
$$ E[||B - \hat{B}||^2]  = E[ (B - \alpha(FB + \varepsilon))^T (B - \alpha(FB + \varepsilon))] = ||(I-\alpha F)B||^2 + E[\varepsilon^T\alpha^T\alpha\varepsilon]$$

Первое слагаемое - квадрат нормы смещения оценки $\hat{B} = \alpha Y$. Заметим, что если невозомжно построить несмещенную оценку, то  $E[||B - \hat{B}||^2]$ может принимать произвольно большие значения. Поэтому оценку следует искать в классе несмещенных (если такие существуют). Условием существование несмещенной оценки является $rank(F) = K$ - условие полного столбцового ранга матрицы $F$.  
В случае полного столбцового ранга  $F^T F$ - невырожденная матрица, и   
(3)  
$$ \hat{\alpha} = (F^TF)^{-1}F^T$$  
а СК оптимальная линейная оценка:  
(4)  
$$ \hat{B} = (F^TF)^{-1}F^TY $$  
носит название оценки Гаусса-Маркова.  
Проверим оптимальность $\hat{B}$. Во-первых, $(I - \hat{\alpha}F)B = (I - (F^TF)^{-1}F^TF)B = 0$, т.е. $\hat{B}$ - несмещенная оценка. Далее, пусть $\bar{B} = (\hat{\alpha} + D)Y$ - другая несмещенная оценка вектора $B$. Из условия несмещеннсоти $\bar{B}$ следует, что 
  
$$ 0 \equiv E[(\bar{B} - B)] = E[(\hat{\alpha} + D)Y - B] = E[\hat{B} + DY - B] = E[DFB] \; \forall B$$  
т.е. матрица D такова, что $DF = 0$. Сначала вычислим $E[||\hat{B}-B||^2]$, а потом сравним это значения критерия с $E[||\bar{B} - B||^2]$  
  
$$E[||\hat{B} - B||^2] = ||(I - \alpha * F)B||^2 + E[\varepsilon^T\hat{\alpha}^T\hat{\alpha}\varepsilon] = tr E[\hat{\alpha}\varepsilon\varepsilon^T\alpha^T = \sigma^2 tr[(F^TF)^{-1}]$$

Вычислим : 
$$ E[||\bar{B} - B||^2] = E[\varepsilon^T(\hat{\alpha} + D)^T(\alpha + D)\varepsilon] = tr E[(\hat{\alpha} + D)(\hat{\alpha} + D)^T] = \sigma^2 tr[(F^TF)^{-1}] + \sigma^2 tr[DD^T] \geqslant E[||\hat{B} - B||^2]$$

т.е. оценка Гаусса-Маркова действительно наилучшая в классе линейных оценок.

#### Замечание №1  
Оценка Гаусса-Маркова может быть построена как в условиях, когда дисперсия шумов в наблюдениях известна, так и при неизвестном значении $\sigma^2$  

#### Замечание №2  
Условие $E[\varepsilon \varepsilon^T] = \sigma^2I$  называется условием **гомоскедастичности** ошибок наблюдений. 

#### Замечание №3
Оценка Гаусса-Маркова оптимальная в классе линейных оценок. Можно ли ее улучшить, если перейти к нелинейным оценкам?


## Связь оценки Гаусса-Маркова и оценок максимального правдоподобия

### Задача №2
В условиях задачи 1 найти ОМП вектора $B$ и дисперсии $\sigma^2$.  
Решение:  
В силу гетероскедастичности шумов в наблюдениях имеем:  
(5)  
$$ L(y, B, \sigma^2) = (2\pi)^{-T/2}(\sigma^2)^{-T/2} exp \{{\frac{1}{2\sigma^2}(y-FB)^T(y-FB)}\} \rightarrow max$$  
Беря логарифм от $L(\cdot)$, задача максимизации преобразуется в задачу:  
(6)  
$$\tilde{L}(y, B, \sigma^2) = \frac{T}{2} ln(\sigma^2) + \frac{1}{2\sigma^2}(y-FB)^T(y-FB) \rightarrow min $$  
и соответствующие нормальные уравнения имеют вид:  
(7)
\begin{equation*}
 \begin{cases}
 \frac{\partial{\tilde{L}}}{\partial{B}} = \frac{1}{2\sigma^2}[-2F^TY + 2F^TFB] = 0\\ 
 \frac{\partial{\tilde{L}}}{\partial{\sigma^2}} = \frac{T}{2}\cdot \frac{1}{\sigma^2}-\frac{1}{2\sigma^4}(y-FB)^T(y - FB) = 0
 \end{cases}
\end{equation*}  

отсюда следует, что ОМП вектора $B$ и дисперсии $\sigma^2$ равна:  
(8)
\begin{equation*}
\begin{cases}
\tilde{B} = \hat{B} = (F^TF)^{-1}F^{T}Y \\
\tilde{\sigma^2} = \frac{1}{T} ||Y-F\tilde{B}||^2
\end{cases}
\end{equation*}


 

### Задача №3

1) Является ли $\tilde{\sigma}^2$ несмещенной оценкой дисперсии - нет  
2) Является ли $\tilde{\sigma}^2$ асимптотически несмещенной оценкой дисперсии - да  
3) Построить насмещенную оценку дисперсии - $\sigma_{н}^2 = \frac{1}{T-K}||Y - F\tilde{B}||^2$  
4) Является ли ОМП $\tilde{B}$ и $\tilde{\sigma}^2$ R-эффективной в случае, если $\sigma^2$ известна? (да)  
5) Является ли ОМП $\tilde{B}$ и $\tilde{\sigma}^2$ R-эффективной в случае, если $\sigma^2$ неизвестна? (нет)

#### Замечание № 4
Из задач МП (5) и (6) следует связь ОМП $\tilde{B}$ и ОМНК $\hat{B}$ - критерии оптимальности разные, но оптимальная оценка - одна.  

#### Замечание № 5
Из решения задачи 3 следует, то в случае гауссовских шумов $\varepsilon$ переход от линейных оценок к нелинейным не повысит качества оценивания. 

### Случай гетероскедастичности шумов
Мы снова рассматриваем модель наблюдений (1'), но $\varepsilon \sim N(0,\sigma^2R)$, причем $R$ - невырожденная матрица, а так же верно условие (2). 
Оценку $\hat{B}$ МНК можно строить, решая одну из следующих оптимизационных задач:  
1)  $E[||B-\bar{B}||^2] \rightarrow min, \; \bar{B} = \bar{\alpha}Y$ - задача СК оптимального линейного оценивания (9)  
2) $(Y - F\bar{B})^T(\sigma^2R)^{-1}(Y - F\bar{B}) \rightarrow min$ - задача взвешенных наименьших квадратов (10)

Решение обеих задач будет одинаковым - оценка Гаусса-Маркова:  
(4')  
$$ \hat{B} = (F^TR^{-1}F)^{-1}F^TR^{-1}Y$$

#### Замечание № 6
Решение задачи (9) может быть сведено к случаю гомоскедастичных шумов, умножив обе части (1') на $R^{-1/2}$:    
  
$$Y' = R^{-1/2}Y = R^{-1/2}F B + R^{-1/2}\varepsilon = \Delta FB + \epsilon$$ - где $\epsilon$ - это гомоскедастичный шум.

## Оценки МНК с линейными ограничениями
Иногда в векторе регрессоров имеется дополнительная априорная информация в виде линейных неравенств: $GB = C$ (11), где $C$ - m мерный вектор столбец (m < K), $G$ - m x K - мерная матрица, причем rank(G) = m. Нужно решить задачу условной оптимизации:  
(12)  
$$ (Y-FB)^{-1}R^{-1}(Y-FB) \rightarrow min, \; B: GB = C$$
Решение:
Так как $G$ - матрица полного строчного ранга, то (11) можно рассматривать как систему линейных алгебраических уравнений, и ее общее решение представимо в виде:  
(13)  
$$ B = G^{+}C + (I - G^{+}G)D$$
где $G^{+} = G^{T}(GG^T)^{-1}$, $D$ - произвольный вектор соответствующей размерности. Будем искать решение оптимизационной задачи (12) в виде $\tilde{B} = \hat{B} + \gamma Y$. Где $\hat{B}$ оценка Гаусса-Маркова (4'). Тогда $\tilde{B}$ должно удовлетворять ограничению (11): $G\tilde{B} = C$.   
$$G(\hat{B} + \gamma Y) = C, \; G\gamma Y = C - G\hat{B}, \; \gamma Y = G^{+}(C-G\hat{B}) + (I - G^{+}G)D$$, где $D$ вектор полжеит отысканию. Т.о. допустимая оценка $\widetilde{B}$ имеет вид $\widetilde{B} = \hat{B} + G^{+}(C-G\hat{B}) + (I - G^{+}G)D$, где $D$ нужно найти. Доказать, что $\tilde{B} = \hat{B} + G^{+}(C - G\hat{B})$ (14).



## Оценка метода нименьших модулей

В гомоскедастичном случае оценка МНК может быть получена с помощью решения следующей оптимизационной задачи:  

$$\sum_{i = 1}^{T}(y_{t_i} - \sum_{k = 1}^{K} f_k(t_i)\beta_k)^2 \rightarrow min$$  

В МНК функция штрафа квадратичная. С одной стороны, это согласуется с ММП в случае, когда ошибки - гауссовские. С другой - оценка МНК очень чувствительная к выбросам, которые появляются в случае гауссовских шумов. 
Пусть шумы в наблюдениях - независимые случайные велчины, имеющие распределение Лапласа с плотностью: 
$$ f(x, \lambda) = \frac{\lambda}{2} exp\{{-\lambda |x|}\}$$  
Параметр $\lambda$ предполагается известным. В этом случае задача оценивания вектора параметров $(\beta_1,..., \beta_k)^T$ сводится к решению оптимизационной задачи:  
$$\sum_{i = 1}^{T} |y_{t_i} - \sum_{k=1}^{K}f_k(t_i)\beta_k| \rightarrow min$$
которая назыается **задачей о наименьших модулях**.  
Данная задача является негладкой, и решить ее можно только численно. Для этого можно использовать итерационный алгоритм, основанный на МНК. 

Шаг 1. 
$$(\beta^{0}_1,..., \beta_K^{0}) = argmin \sum_{i = 1}^{T} (y_{t_i} - \sum_{k = 1}^{K}f_k(t_i)\beta_k)^2$$
Шаг 2. (повторяется в цикле)  
Пусть известна оценка $(\beta^{n-1}_1,..., \beta_K^{n-1})$ на предыдущей итерации. Тогда :
$$(\beta^{n}_1,..., \beta_K^{n}) = argmin \sum_{i = 1}^{T} \frac{1}{\sigma_i^{n-1}} (y_{t_i} - \sum_{k = 1}^{K}f_k(t_i)\beta_k)^2$$
где $\sigma_i^{n-1} = |y_{t_i} - \sum_{k = 1}^{K}f_k(t_i)\beta_k^{n-1}|$  
Выход из цикла : $||B^n - B^{n-1}|| \leqslant \delta$  
Данный алгоритм называется алгоритмом вариационно-взвешенных наименьших квадратов
